```python
	# upload the file
from google.colab import files
uploaded = files.upload()
	# to read Dataset
import pandas as pd
	#file Name dataset_salary.csv
dataset=pd.read_csv("dataset_salary.csv")
	# to separate the column 0 , and used value to cancle the name of column
	# use [0] square brackets because x can be a matrix
x=dataset.iloc[:,[0]].values
	# to separate the column 1
y=dataset.iloc[:,1].values
	# ŸçSpllitting training testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
	# Training .... algorithm
from sklearn.linear_model import LinearRegression
	# Class ....object
regressor=LinearRegression()
	# Fit function
regressor.fit(x_train,y_train)
	# testing
ypred=regressor.predict(x_test)
	# Evaluation
from sklearn.metrics import mean_absolute_error
print (mean_absolute_error(y_test,ypred))
from os import copy_file_range
	# Slop
A=regressor.coef_
	# Intersection
b=regressor.intercept_
print ("y=",A,"x + ",b)
print ("the salary of the employee is :" , regressor.predict ([[10.5]]))
print ("the salary of the employee is :" , regressor.predict ([[12.5]]))
import matplotlib.pyplot as plt   #visualization train
plt.scatter(x_train,y_train,color='red')
```
