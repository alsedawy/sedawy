```python
# Add the module optimizer, datasets, layers (dense - conv2d - maxpool2d - flatten)
from tensorflow.keras import optimizers, datasets, layers
import tensorflow as tf
# Load Dataset (mnist)
(train_x, train_y),(test_x, test_y)=datasets.mnist.load_data()
#Check the shape of data
train_x.shape
# Reshape it to become 3 D
train_x=train_x.reshape(60000,28,28,1)
test_x=test_x.reshape(10000,28,28,1)
train_x.shape
print(train_x[0])
# Normalization by devided to 255 to become from ( 0 to 1)
train_x=train_x.astype('float32')/255
test_x=test_x.astype('float32')/255
print(train_x[0])
print(train_y[0])
# To View image
import matplotlib.pyplot as plt
plt.imshow(train_x[0])
# conver as vector by one hot encoding
train_y=tf.keras.utils.to_categorical(train_y)
test_y=tf.keras.utils.to_categorical(test_y)
train_y[0]
from tensorflow.keras import activations, Sequential
# Build of layers
model1=Sequential()
# layer(conv2D) filter(5*5)32time/image strdid(1) padding(same) activation(ReLu) shap(3D)
model1.add(layers.Conv2D(filters=32,
    kernel_size=5,strides=1, padding="same",
    activation=tf.nn.relu, input_shape=(28,28,1)))
# Layed(Maxpool2D) window(2*2) strides=(2*2) Padding(valid)
model1.add(layers.MaxPool2D(pool_size=(2,2),strides=(2,2),padding='valid'))
# Layer(conv2D) filter(3*3)64time/image strdid(1) padding(same) activation(ReLu)
model1.add(layers.Conv2D(filters=64,
    kernel_size=3,strides=1, padding="same",
    activation=tf.nn.relu))
# Layer(Maxpool2D) window(2*2) strides=(2*2) Padding(valid)
model1.add(layers.MaxPool2D(pool_size=(2,2),strides=(2,2),padding='valid'))
# Layer(Flatten)
model1.add(layers.Flatten())
# Layer(dense) Nurel(128) activation(ReLu)
model1.add(layers.Dense(units=128, activation=tf.nn.relu))
# Output Layer(dense) Nurel(10) activation(Soft Max)
model1.add(layers.Dense(units=10,activation=tf.nn.softmax))
# compilation -- in Multi class [loss="categorical_crossentropy]
model1.compile(optimizer='adam', loss="categorical_crossentropy",metrics=['accuracy'])
# Fit validation(test_x,test_y)
hist=model1.fit(x=train_x, y=train_y,epochs=100,validation_data=(test_x,test_y),
      batch_size=128)
model1.summary()
import matplotlib.pyplot as plt
# Summarize history for accurance
plt.plot(hist.history['accuracy'])
plt.plot(hist.history['val_accuracy'])
plt.title('model1 accturacy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train','validation'],loc='upper left')
plt.show
# Summarize history for Loss
plt.plot(hist.history['loss'])
plt.plot(hist.history['val_loss'])
plt.title('model1 loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train','validation'],loc='upper left')
plt.show
```
