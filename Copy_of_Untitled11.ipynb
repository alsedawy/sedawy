```python
# ==========================
# Collab-ready: كل الكود في خلية واحدة
# ==========================
# تثبيت الحزم (قد يستغرق تنزيل النماذج وقتًا)
!pip install -q sentence-transformers scikit-learn matplotlib seaborn requests rank_bm25 transformers chardet

# ==========================
# الواردات
# ==========================
import os, sys, time
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from rank_bm25 import BM25Okapi
from collections import defaultdict
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
from transformers import pipeline
import chardet
from google.colab import files

# ==========================
# 1) رفع الملف
# ==========================
print("اختر ملف IT_Q&A.txt من جهازك لرفعه إلى Colab (ستظهر نافذة رفع الملفات)...")
uploaded = files.upload()
if not uploaded:
    raise RuntimeError("لم يتم رفع أي ملف. أعد تشغيل الخلية وارفع ملف IT_Q&A.txt.")

filename = list(uploaded.keys())[0]
print("Loaded file:", filename)

# ==========================
# 2) كشف الترميز وقراءة الملف بأمان
# ==========================
with open(filename, "rb") as f:
    raw = f.read()

det = chardet.detect(raw)
encoding = det.get("encoding") or "utf-8"
print("Detected encoding:", encoding)

# اقرأ النص مع تجاهل الأخطاء الطفيفة إن وجدت
with open(filename, "r", encoding=encoding, errors="ignore") as f:
    fetched_content = f.read()

print("File loaded. Length:", len(fetched_content))

# ==========================
# 3) تقسيم الأسئلة والأجوبة (داعم للعربية والإنجليزية)
# ==========================
lines = [line.strip() for line in fetched_content.splitlines() if line.strip()]

qa_pairs = []
current_question = None
current_answer = []

question_starts = (
    "what","how","do","are","is","why","where","when","which",
    "ما","ماذا","هل","لماذا","أين","متى","كيف","من","ماهي","ما هي"
)

for line in lines:
    parts = line.split()
    first_token = parts[0].lower().strip(":،؟") if parts else ""
    is_question = line.endswith("?") or (first_token in question_starts)
    if is_question:
        if current_question and current_answer:
            qa_pairs.append((current_question.strip(), " ".join(current_answer).strip()))
        current_question = line
        current_answer = []
    else:
        current_answer.append(line)

if current_question and current_answer:
    qa_pairs.append((current_question.strip(), " ".join(current_answer).strip()))

print("Total Q/A pairs:", len(qa_pairs))
if len(qa_pairs) == 0:
    raise RuntimeError("qa_pairs فارغة — تحقق من شكل الملف أو محتواه.")

# (اختياري) طباعة أول 3 أزواج للتأكد
for i, (q,a) in enumerate(qa_pairs[:3]):
    print(f"\nQA[{i}]\nQ: {q}\nA_preview: {a[:200]}...")

# ==========================
# 4) إعداد نماذج التضمين و GPT-2 (إن أمكن)
# ==========================
print("\nLoading embedding model (all-MiniLM-L6-v2). This may take a moment...")
EMBEDDING_MODEL = SentenceTransformer('all-MiniLM-L6-v2')

# حاول تحميل GPT-2 المحلي — إن فشل، الكود سيعمل بدون التوليد (سيستخدم إجابة استخلاصية)
LOCAL_GENERATOR = None
try:
    print("Loading local text-generation pipeline (gpt2)...")
    LOCAL_GENERATOR = pipeline("text-generation", model="gpt2", tokenizer="gpt2", device= -1)  # cpu
    print("Local generator loaded.")
except Exception as e:
    print("تحذير: لم يتم تحميل GPT-2 المحلي (قد يكون بسبب قيود الشبكة/الذاكرة). التوليد سيكون موصولًا بخطة بديلة (استخلاص).")
    LOCAL_GENERATOR = None

# ==========================
# 5) دوال المساعدة للاسترجاع/التوليد/التقييم
# ==========================
def get_embedding(text):
    return EMBEDDING_MODEL.encode([text])[0]

def retrieve_bm25(query, top_k=5):
    global qa_pairs
    corpus = [(q + " " + a).strip() for q, a in qa_pairs if (q + " " + a).strip()]
    if not corpus:
        return []
    tokenized = [doc.split() for doc in corpus]
    if not tokenized:
        return []
    bm25 = BM25Okapi(tokenized)
    scores = bm25.get_scores(query.split())
    ranked = sorted(enumerate(scores), key=lambda x: x[1], reverse=True)
    return ranked[:top_k]

def semantic_search(query, qa_pairs_local, top_k=5):
    corpus = [q + " " + a for q, a in qa_pairs_local]
    emb = EMBEDDING_MODEL.encode(corpus)
    q_emb = get_embedding(query)
    scores = cosine_similarity([q_emb], emb)[0]
    ranked = sorted(enumerate(scores), key=lambda x: x[1], reverse=True)
    return ranked[:top_k]

def rewrite_query(query):
    return [
        f"{query} How to solve?",
        f"Steps to {query.lower()}?",
        f"{query} what are the steps to fix it?"
    ]

def rrf_fuse(rank_lists, k=5, c=60):
    scores = defaultdict(float)
    for rlist in rank_lists:
        for rank, doc_id in enumerate(rlist):
            scores[doc_id] += 1.0 / (c + rank + 1)
    fused = sorted(scores.items(), key=lambda x: x[1], reverse=True)
    return [doc for doc, _ in fused[:k]]

def rerank_semantic_over_bm25(query, qa_pairs_local, coarse_k=6, final_k=3):
    bm25_hits = retrieve_bm25(query, top_k=coarse_k)
    coarse_ids = [i for i, _ in bm25_hits]
    if not coarse_ids:
        return []
    docs = [(qa_pairs_local[i][0] + " " + qa_pairs_local[i][1]) for i in coarse_ids]
    emb_docs = EMBEDDING_MODEL.encode(docs)
    q_emb = get_embedding(query)
    scores = cosine_similarity([q_emb], emb_docs)[0]
    reranked = sorted(zip(coarse_ids, scores), key=lambda x: x[1], reverse=True)
    return [idx for idx, _ in reranked[:final_k]]

def build_rag_prompt(question, contexts,
                     instruction="Answer the question using ONLY the context below. If not present, say 'I don't know from the provided context.'"):
    ctx_block = "\n\n---\n\n".join(contexts)
    return f"{instruction}\n\nContext:\n{ctx_block}\n\nQuestion: {question}\nAnswer:"

def local_generate(prompt, max_tokens=120, temperature=0.2):
    if LOCAL_GENERATOR is None:
        return None
    try:
        out = LOCAL_GENERATOR(prompt, max_length=max_tokens, temperature=temperature, do_sample=True, return_full_text=False)
        if isinstance(out, list) and len(out) > 0 and 'generated_text' in out[0]:
            return out[0]['generated_text'].strip()
        # transformers different return
        return out[0].get('generated_text', None) if isinstance(out, list) else None
    except Exception as e:
        return None

# Evaluation metrics
def token_set(text): return set(text.lower().split())
def overlap_f1(pred, gold):
    p, g = token_set(pred), token_set(gold)
    if not p and not g: return 1.0
    if not p or not g: return 0.0
    prec = len(p & g) / len(p) if p else 0
    rec = len(p & g) / len(g) if g else 0
    if prec + rec == 0: return 0.0
    return 2 * prec * rec / (prec + rec)

def hallucination_rate(pred, supporting_texts):
    p_toks = set(pred.lower().split())
    sup_toks = set()
    for s in supporting_texts: sup_toks |= set(s.lower().split())
    if not p_toks: return 0.0
    not_in = sum(1 for t in p_toks if t not in sup_toks)
    return not_in / len(p_toks)

def fidelity(pred, supporting_texts):
    p_toks = set(pred.lower().split())
    sup_toks = set()
    for s in supporting_texts: sup_toks |= set(s.lower().split())
    if not p_toks: return 0.0
    in_count = sum(1 for t in p_toks if t in sup_toks)
    return in_count / len(p_toks)

# ==========================
# 6) دالة تشغيل التقييم (تشغيل جميع الطرق)
# ==========================
def run_evaluation_with_hf(test_questions, method="baseline", top_k_retr=3):
    results = []
    for (q_gold, a_gold) in test_questions:
        query = q_gold
        # retrieval according to method
        if method == "baseline":
            hits = retrieve_bm25(query, top_k=top_k_retr)
            retrieved_idxs = [i for i, _ in hits]
        elif method == "semantic":
            hits = semantic_search(query, qa_pairs, top_k=top_k_retr)
            retrieved_idxs = [i for i, _ in hits]
        elif method == "hybrid_fusion":
            bm25_hits = retrieve_bm25(query, top_k=top_k_retr)
            semantic_hits = semantic_search(query, qa_pairs, top_k=top_k_retr)
            bm25_rank = [i for i, _ in bm25_hits]
            sem_rank = [i for i, _ in semantic_hits]
            fused = rrf_fuse([bm25_rank, sem_rank], k=top_k_retr)
            retrieved_idxs = fused
        elif method == "reranking":
            retrieved_idxs = rerank_semantic_over_bm25(query, qa_pairs, coarse_k=6, final_k=top_k_retr)
        elif method == "fusion":
            variants = rewrite_query(query)
            rank_lists = []
            hits_orig = retrieve_bm25(query, top_k=top_k_retr)
            rank_lists.append([i for i, _ in hits_orig])
            for v in variants:
                hits = retrieve_bm25(v, top_k=top_k_retr)
                rank_lists.append([i for i, _ in hits])
            fused = rrf_fuse(rank_lists, k=top_k_retr)
            retrieved_idxs = fused
        else:
            retrieved_idxs = [i for i,_ in retrieve_bm25(query, top_k=top_k_retr)]

        # contexts
        supporting_texts = [qa_pairs[int(idx)][1] for idx in retrieved_idxs if int(idx) < len(qa_pairs)]
        prompt = build_rag_prompt(query, supporting_texts) if supporting_texts else build_rag_prompt(query, [""])

        # generation
        gen = local_generate(prompt, max_tokens=120)
        if gen is None or len(gen.strip()) < 3:
            pred = " ".join(supporting_texts[:3]) if supporting_texts else ""
            used_ctx = supporting_texts
        else:
            pred = gen
            used_ctx = supporting_texts

        f1 = overlap_f1(pred, a_gold)
        hr = hallucination_rate(pred, used_ctx)
        fid = fidelity(pred, used_ctx)

        results.append({
            "query": query, "gold": a_gold, "pred": pred,
            "overlap_f1": f1, "hallucination_rate": hr, "fidelity": fid
        })

    # averages
    num_results = len(results)
    avg_f1 = sum(r["overlap_f1"] for r in results) / num_results if num_results else 0
    avg_hr = sum(r["hallucination_rate"] for r in results) / num_results if num_results else 0
    avg_fid = sum(r["fidelity"] for r in results) / num_results if num_results else 0

    return results, {"avg_f1": avg_f1, "avg_hallucination_rate": avg_hr, "avg_fidelity": avg_fid}

# ==========================
# 7) مجموعة الاختبار (أمثلة عربية — عدّل كما تريد)
# ==========================
test_set = [
    ("ما هي المسؤوليات الأساسية لمهندس الدعم الفني؟", "The work of a technical support engineer is to maintain and monitor the computers and the networks of an organization..."),
    ("صِف منهجيتك في تشخيص المشكلات وحلها.", "This question checks your approach to identifying a problem and finding its solution..."),
    ("ما هو الفرق الجوهري بين SDK و API؟", "An API (Application Programming Interface) is a set of instructions, protocols, and tools for building software applications..."),
    ("اشرح ذاكرة التخزين المؤقت (Cache).", "Cache memory is a small, fast memory buffer used by the central processing unit (CPU)..."),
    ("ما هي الخطوات لحل مشكلة الوصول إلى ملف في محرك مشترك؟", "The first step is to check if the network connection is working correctly...")
]

# ==========================
# 8) تشغيل المقارنات لجميع الطرق
# ==========================
methods = ['baseline', 'reranking', 'fusion', 'semantic', 'hybrid_fusion']
names = ['Baseline', 'Reranking', 'RRF (BM25)', 'Semantic', 'Hybrid']

print("\nRunning evaluations (this may take time depending on embedding/model loads)...")
metrics = []
all_results = {}

for m in methods:
    print(f"→ Running method: {m}")
    res, avg = run_evaluation_with_hf(test_set, method=m, top_k_retr=3)
    metrics.append(avg)
    all_results[m] = res
    print(f"   Done. Metrics: {avg}")

# ==========================
# 9) عرض النتائج في DataFrame ورسم
# ==========================
df = pd.DataFrame({
    "Method": names,
    "F1": [m["avg_f1"] for m in metrics],
    "Hallucination": [m["avg_hallucination_rate"] for m in metrics],
    "Fidelity": [m["avg_fidelity"] for m in metrics],
})

print("\nSummary table:")
print(df)

plt.figure(figsize=(16,5))
sns.set_style("whitegrid")

plt.subplot(1,3,1)
sns.barplot(x="Method", y="F1", data=df)
plt.ylim(0,1)
plt.title("F1 Score")

plt.subplot(1,3,2)
sns.barplot(x="Method", y="Hallucination", data=df)
plt.ylim(0,1)
plt.title("Hallucination Rate")

plt.subplot(1,3,3)
sns.barplot(x="Method", y="Fidelity", data=df)
plt.ylim(0,1)
plt.title("Fidelity")

plt.tight_layout()
plt.show()

# ==========================
# 10) حفظ النتائج إن رغبت (اختياري)
# ==========================
df.to_csv("rag_evaluation_summary.csv", index=False)
print("\nSaved summary to rag_evaluation_summary.csv")
```
